```python
from transformers import pipeline

emotion_model = pipeline("text-classification", model="j-hartmann/emotion-english-distilroberta-base",
    return_all_scores=False
)
```

```python
label_to_emoji = {
    "sadness": "😢",
    "joy": "😂",
    "anger": "😡",
    "surprise": "😲",

}
def predict_emoji(text: str) -> str:
    text = text.strip()
    if not text:
        return "😐"

    result = emotion_model(text)[0]
    label = result["label"].lower()

    emoji = label_to_emoji.get(label, "😐")
    return emoji
```

```python
tests = [
    "I feel terrible and want to cry",
    "LOL that was so funny",
    "Why did you do that? I'm furious",
    "Wait, what just happened?!",
    "I really love this",
    "I'm scared about tomorrow"
]

for t in tests:
    print(t, "->", predict_emoji(t))
```

```python
import gradio as gr

 # بناء وجهة نفاعلية
iface = gr.Interface(
    #الداله اللي تشتغل لما المستخدم يدخل النص
    fn=predict_emoji,
    inputs=gr.Textbox(lines=2, placeholder="Enter an English phrase..."), outputs="text",
    title="Transformer Emoji Predictor",
    description="Type a phrase and get an emoji: 😢 😂 😡 😲"
)

iface.launch()
```
